## Top 10 ODI Teams - Men

Top 10 ODI teams in mens cricket along with the records for matches, points and rating.

```python
import requests
import lxml.html as lh
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url='https://www.icc-cricket.com/rankings/mens/team-rankings/odi'
#Create a handle, page, to handle the contents of the website
page = requests.get(url)
#Store the contents of the website under doc
doc = lh.fromstring(page.content)
#Parse data that are stored between <tr>..</tr> of HTML
tr_elements = doc.xpath('//tr')
```

```python
#Check the length of the first 12 rows
[len(T) for T in tr_elements[:12]]
```

```python
tr_elements = doc.xpath('//tr')
#Create empty list
col=[]
i=0
#For each row, store each first element (header) and an empty list
for t in tr_elements[0]:
    i+=1
    name=t.text_content()
    print (i,name)
    col.append((name,[]))
```

```python
#Since out first row is the header, data is stored on the second row onwards
for j in range(1,len(tr_elements)):
    #T is our j'th row
    T=tr_elements[j]
    
    #If row is not of size 10, the //tr data is not from our table 
    if len(T)!=5:
        break
    
    #i is the index of our column
    i=0
    
    #Iterate through each element of the row
    for t in T.iterchildren():
        data=t.text_content() 
        #Check if row is empty
        if i>0:
        #Convert any numerical value to integers
            try:
                data=int(data)
            except:
                pass
        #Append the data to the empty list of the i'th column
        col[i][1].append(data)
        #Increment i for the next column
        i+=1
```

```python
[len(C) for (title,C) in col]
```

```python
Dict={title:column for (title,column) in col}
df=pd.DataFrame(Dict)
df
```

```python
df_new=df
```

```python
df_new.drop(df_new.columns[1], axis=1, inplace=True)
df_new
```

```python
page6=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
soup7=BeautifulSoup(page6.content)
soup7
team_name=soup7.find_all('span',class_="u-hide-phablet")
team_name
```
