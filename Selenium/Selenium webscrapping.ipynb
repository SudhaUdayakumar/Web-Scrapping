<B>Q2: Write a python program to scrape data for Data Scientist Job position inBangalore location. You have to scrape the job-title, job-location,company_name, full job-description. You have to scrape first 10 jobs data.<B>

```python
url1= 'https://www.naukri.com/'
driver.get(url1)
```

```python
search_job.send_keys('Data Scientist')
```

```python
search_1=driver.find_element_by_xpath('//button[@class="btn"]')
search_1.click()
```

```python
titles=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
titles
```

```python
job_titles=[]
for i in titles[:10]:
    job_titles.append(i.text)
job_titles
```

```python
company=driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
company
```

```python
company_name=[]
for i in company[:10]:
    company_name.append(i.text)
company_name
```

```python
job_location=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']//span")
job_location
```

```python
locations=[]
for i in job_location[:10]:
    locations.append(i.text)
locations
```

```python
urls=[] 
for i in driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']"):
    urls.append(i.get_attribute("href"))
urls
```

```python
job_description=[] #defining variables 
```

```python
for i in urls[:10]:
    driver.get(i)
    time.sleep(3)
        
    #fetching job description
    try:
        job_desc=driver.find_element_by_xpath("//section[@class='job-desc']")
        job_description.append(job_desc.text)
    except NoSuchElementException:
        job_description.append('-')
```

```python

job_description
```
