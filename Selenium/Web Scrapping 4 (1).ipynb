## Q1 Wikipedia: Top viewed YouTube videos

<b>1. Scrape the details of most viewed videos on YouTube from Wikipedia:
    
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/
    
You need to find following details:
    
A) Rank
    
B) Name
    
C) Artist
    
D) Upload date
    
E) Views
</b>

```python
url='https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver.get(url)  
```

```python
#creating empty lists

rank=[]
name=[]
artist=[]
upload_date=[]
views=[]

#rank
video_rank=driver.find_elements_by_xpath('/html/body/div[3]/div[3]/div[5]/div[1]/table[3]/tbody/tr/td[1]')
for i in video_rank:
    rank.append(i.text)

#name
video_name=driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr/td[2]')  
for i in video_name:   
    name.append(i.text)
    
#artist
video_artist=driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr/td[3]')  
for i in video_artist:   
    artist.append(i.text)
    
#upload date
video_date=driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr/td[5]')  
for i in video_date:   
    upload_date.append(i.text)
    
#views
video_views=driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr/td[4]')  
for i in video_views:   
    views.append(i.text)
    
```

```python
#printing the length of all elements
print(len(rank))
print(len(name))
print(len(artist))
print(len(upload_date))
print(len(views))
```

```python
YouTube=pd.DataFrame({})   #creating a dataframe
YouTube['Rank']=rank
YouTube['Name of video']=name
YouTube['Artist/Uploader']=artist
YouTube['Upload Date']=upload_date
YouTube["Views(in billion)"]=views
YouTube
```

B) Repository description

D) Language used
    
Note: - From the home page you have to click on the trending option from Explore menu through code</b>

```python
url4='https://github.com/'
driver.get(url4)

explore=driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/summary').click()  #clicking on explore
trend=driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul[2]/li[3]/a').click()  #clicking on trending
```

```python
from selenium.common.exceptions import NoSuchElementException    #importing NoSuchElementException
```

```python
#scrapping urls
urls=[]
for i in driver.find_elements_by_xpath('//*[@id="js-pjax-container"]/div[3]/div/div[2]/article/h1/a'):
    urls.append(i.get_attribute("href"))
```

```python
#creating empty lists
rep_title=[]
rep_desc=[]
contributors=[]
language=[]

time.sleep(2)
#scrapping rest of required data
for i in urls:
    driver.get(i)
    
    #repository title
    try:
        repos_title=driver.find_element_by_xpath('//h1[@class=" d-flex flex-wrap flex-items-center break-word f3 text-normal"]')
        rep_title.append(repos_title.text.replace('\n',' '))
    except NoSuchElementException:
        rep_title.append('-')
        
    #repository description
    try:
        repos_desc=driver.find_element_by_xpath('//*[@id="repo-content-pjax-container"]/div/div[2]/div[2]/div/div[1]/div/p')
        rep_desc.append(repos_desc.text)
    except NoSuchElementException:
        rep_desc.append('Not Found')
        
    #contributors count
    try:
        cont_count=driver.find_element_by_xpath('//*[@id="repo-content-pjax-container"]/div/div[2]/div[2]/div/div[5]/div/h2/a/span')
        contributors.append(cont_count.text)
    except NoSuchElementException:
        contributors.append('Not found')
    #language used
    try:
        lang=driver.find_element_by_xpath("//div[@class='BorderGrid-row']/div/ul[@class='list-style-none']")
        language.append(lang.text.replace('\n',' '))
    except NoSuchElementException:
        language.append('Not Found')
        
#printing length of values
print(len(rep_title),len(rep_desc),len(contributors),len(language))
```

```python
GitHub=pd.DataFrame({'Respository Title':rep_title, 'Repository Description':rep_desc, 'Contributors count':contributors,'Lanugages':language})
GitHub  #creating a dataframe
```

```python
#creating empty lists 
name2=[]
desg=[]
company=[]
skills=[]
loc=[]


time.sleep(2)


#name 
rec_name=driver.find_elements_by_xpath("//span[@class='fl ellipsis']")
for i in rec_name:
    name2.append(i.text)
    
#designation 
rec_desg=driver.find_elements_by_xpath("//span[@class='ellipsis clr']")
for i in rec_desg:
    desg.append(i.text)
    
#company
rec_comp=driver.find_elements_by_xpath("//a[@class='ellipsis'][2]")
for i in rec_comp:
    company.append(i.get_attribute('title'))
    
#skills they hire for 
rec_skills=driver.find_elements_by_xpath("//div[@class='hireSec highlightable']")
for i in rec_skills:
    skills.append(i.text)
    
#getting recruiters url
urls1=[]
rec_urls=driver.find_elements_by_xpath("//a[@class='ellipsis'][1]")
for i in rec_urls:
    urls1.append(i.get_attribute("href"))
    
for i in urls1:
    driver.get(i)
    
    #getting location 
    try:
        rec_loc=driver.find_element_by_xpath("/html/body/div[2]/div[1]/div[1]/div[2]/div[5]/a")
        loc.append(rec_loc.text)
    except NoSuchElementException:
        loc.append('-')
#printing length of values
print(len(name2),len(desg),len(company),len(loc),len(skills))
```

## Q10: UCI machine learning repositories.

Url = https://archive.ics.uci.edu/
    
You have to find the following details:
    
A) Dataset name

B) Data type

C) Task

D) Attribute type

E) No of instances

F) No of attribute

G) Year

Note: - from the home page you have to go to the Show All Dataset page through code.

```python
url9=' https://archive.ics.uci.edu/'
driver.get(url9)

time.sleep(2)
#clicking on 'View all datasets'
data=driver.find_element_by_xpath("/html/body/table[1]/tbody/tr/td[2]/span[2]/a")

driver.get(data.get_attribute('href'))
```

```python
#creating empty lists 
dataset=[]
data_type=[]
task=[]
attribute=[]
instances=[]
year1=[]

#scrapping required data 
time.sleep(3)

#dataset name
data_name=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td/table/tbody/tr/td[2]/p/b/a")
for i in data_name: 
    dataset.append(i.text)
    
#datatype
datatype=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]/p")
for i in datatype[1:]:
    data_type.append(i.text)
    
#task
tasks=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]/p')
for i in tasks[1: ]:
    task.append(i.text)

#attribute type
att_type=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]/p")
for i in att_type [1:]:
    attribute.append(i.text)

#instances
inst=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]/p")
for i in inst[1:]:
    instances.append(i.text)
    
#year
data_year=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]/p")
for i in data_year[1: ]:
    year1.append(i.text)
```

```python
#printing length of values 
print(len(dataset),len(data_type),len(tasks),len(attribute),len(instances),len(year1))
```

```python
Repositories=pd.DataFrame({}) #creating a dataframe
Repositories['Dataset Name']=dataset
Repositories['Data Type']=data_type
Repositories['Tasks']=task
Repositories['Attribute type']=attribute
Repositories['Instances']=instances 
Repositories['Year']=year1
Repositories
```
